```python
#Name: Adewale Kasim ID: KAS20505831
#This is to prompts the User to Input the Size of the Environment and the locations of where the dirts 
def create_environment():
    while True:
        size_str = input("Enter the size of the environment = N (3 or 4 or 5): ")
        try:
            size = int(size_str)
            if size < 3 or size > 5:
                raise ValueError
            break
        except ValueError:
            print("Invalid Size, please enter a Positive Integer between 3 and 5")
    location_of_dirts = []
    for i in range(size):
        for j in range(size):
            if input(f"Is there Dirt at position? ({i+1},{j+1})? (Yes/No)") == "Yes":
                location_of_dirts.append((i, j))
    return size, location_of_dirts

#This is to display the eviroment in a table format but to use this we need to install the php install prettytable in the system taminal
from prettytable import PrettyTable

#This is to display the current state of the Environment, that also includes the locations of the dirts
def display_environment(size, location_of_dirts, cleaner_position=None):
    table = PrettyTable()
    for i in range(size):
        row = []
        for j in range(size):
            if cleaner_position == (i, j):
                row.append("Agent")
            elif (i, j) in location_of_dirts:
                row.append("Dirt")
            else:
                row.append("Clean")
        table.add_row(row)
    print(table)
    print()


#This is to  move the Cleaner based on this inputs (Left, Right, Up, Down, Or Suck)
def move_cleaner(current_position, action, size, location_of_dirts):
    if action == "Left":
        new_position = (current_position[0], max(current_position[1]-1, 0))
    elif action == "Right":
        new_position = (current_position[0], min(current_position[1]+1, size-1))
    elif action == "Up":
        new_position = (max(current_position[0]-1, 0), current_position[1])
    elif action == "Down":
        new_position = (min(current_position[0]+1, size-1), current_position[1])
    elif action == "Suck":
        if current_position in location_of_dirts:
            location_of_dirts.remove(current_position)
        new_position = current_position
    else:
        raise ValueError("Invalid Action")
    return new_position

#This to check if all dirts has been cleaned in the Environment
def check_goal_state(location_of_dirts):
    return len(location_of_dirts) == 0


size, location_of_dirts = create_environment()
cleaner_position_str = input("Enter the starting position of the Cleaner (in the format 'row,col': ")
cleaner_position = tuple(map(int, cleaner_position_str.split(",")))
display_environment(size, location_of_dirts, cleaner_position)
#This is loop function of the Main Program
while not check_goal_state(location_of_dirts):
    action = input("Enter action for the Cleaner (Left/Right/Up/Down/Suck): ")
    cleaner_position = move_cleaner(cleaner_position, action, size, location_of_dirts)
    display_environment(size, location_of_dirts, cleaner_position)
print("All Dirt has been cleaned!")
```
